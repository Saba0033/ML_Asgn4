# Inference

ResNet checkpoint -> predict test.csv -> submission.csv

```python
import sys
sys.path.append(".")

import os
import torch
import pandas as pd
from src.models import build_model
from src.data import make_kaggle_test_loader
from src.engine import get_device

best_arch = "ResNet"
ckpt = torch.load(f"models/{best_arch}_best.pt", map_location="cpu")
model = build_model(ckpt["architecture"], **ckpt["model_cfg"])
model.load_state_dict(ckpt["state_dict"])
device = get_device()
model.to(device).eval()

preds = []
with torch.no_grad():
    for x in make_kaggle_test_loader(batch_size=256):
        preds.append(model(x.to(device)).argmax(1).cpu())

os.makedirs("submissions", exist_ok=True)
pd.Series(torch.cat(preds).numpy()).to_csv("submissions/submission.csv", index=False, header=False)
print("wrote submissions/submission.csv", len(torch.cat(preds)))
!head submissions/submission.csv
```
